```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(4, 15, 100)
y = np.sin(x)
```

```python
x
```

```python
y
```

```python
plt.figure(figsize=(5, 3))
plt.plot(x, y, label="Sin Wave", linestyle="-",linewidth=2, color="blue")
plt.xlabel("X values")
plt.ylabel("Y Value")
plt.title("Sin Wave")
plt.legend()
plt.grid(True)
plt.show()
```

```python
category = ["A", "B", "C", "D", "E"]
value = [33, 54, 42, 51, 29]
```

```python
plt.figure(figsize=(5, 3))
plt.bar(category, value, alpha=0.7, edgecolor="black",
        color=["#555040","#555100","#555060","#555080","#555020"])
plt.xlabel("Categories")
plt.ylabel("Values")
plt.title("Comparing Categories")
plt.grid(True)
plt.show()
```

```python
x = np.random.rand(100)
y = np.random.rand(100)
color = np.random.rand(100)
size = 1000 * np.random.rand(100)
```

```python

plt.scatter(x, y, c=color, s=size, alpha=0.7,cmap="viridis")
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.title("Scatter Plot")
plt.colorbar(label="Color Intensity")
plt.grid(True)
plt.show()
```

```python
dice = np.array([1, 2, 3, 4, 5, 6])  # Dice
sample = np.random.choice(dice, 4) # Rolling dice 4 time in a row and getting values
print("Output of 4 samples", sample)
print("Mean of above 4 samples", sample.mean())
```

```python
samples_1000 = [np.random.choice(dice, 4).mean() for i in range(1000)]
samples_1000
```

```python

plt.hist(samples_1000, alpha=0.7, edgecolor="red", bins=100, histtype="barstacked")
plt.show()
```

```python
data = np.random.randn(10000)


plt.hist(data, alpha=0.7, edgecolor="black", bins=100, histtype="stepfilled",
        color="r")
plt.xlabel("Values")
plt.ylabel("frequency")
plt.title("Histogram")
plt.grid(True)
plt.show()
```

```python
sizes = np.random.randint(20, 89, 4)
labels = ["East", "West", "South", "North"]


plt.pie(sizes, labels=labels, colors=["gold", "lightgreen", "cyan", "lightcoral"],
        autopct="%1.1f%%", explode=[0.01, 0.1, 0.01, 0.01])  # F= float, 0.2 = according to east, west
plt.title("Piechart")
plt.show()
```

```python
data = np.random.normal(loc=0, scale=1,size=(100,))


plt.boxplot(data, patch_artist=True, vert=False)
plt.title("Boxplot")
plt.show()
```

```python
angles = np.linspace(0, 2*np.pi, 100)
radii = np.random.rand(100)
```

```python
plt.figure(figsize=(5,4))
ax = plt.subplot(111, polar=True)
ax.plot(angles, radii, color="blue", linewidth=0.2)
ax.fill(angles, radii, color="blue", alpha=0.3)
plt.title("Polar Plot")
plt.show()
```

```python
matrix = np.random.rand(5,5)
plt.figure(figsize=(5, 4))
plt.imshow(matrix, cmap="plasma")
plt.title("heatmap")
plt.colorbar(label="Color Intenaity")
plt.show()
```

```python
overs = np.arange(1, 21)
ind = np.random.randint(1, 24, 20)
nz = np.random.randint(1, 24, 20)
```

```python
ind_total = [0]
for i in ind:
    ind_total.append(i+ind_total[-1])

nz_total = [0]
for i in nz:
    nz_total.append(i+nz_total[-1])
```

```python
plt.figure(figsize=(10, 3))

plt.subplot(121)   #121=roe,column
plt.plot(overs, nz_total[1:], label="India")
plt.plot(overs, ind_total[1:], label="New Zealand")
plt.legend()

plt.subplot(122)
plt.bar(overs, nz, label="India")
plt.bar(overs, ind, label="New Zealand")
plt.legend()
plt.show()
```
