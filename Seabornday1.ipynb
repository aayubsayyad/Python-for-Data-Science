```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings("ignore")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
# from matlotlib import pyplot as plt
```

```python
weight = [62, 64, 69, 75, 68, 66, 65, 71, 76, 73]
height = [120, 136, 148, 175, 137, 165, 154, 172, 200, 187]
```

```python
%matplotlib inline
plt.figure(figsize=(5,4))
plt.xlabel("Height")
plt.ylabel("Weight")
plt.title("Height vs Weight")
sns.scatterplot(x = height, y = weight)
plt.show()
```

```python
gender = ["Female", "Female", "Female", "Female",
          "Male", "Male", "Male"]
```

```python
plt.figure(figsize=(4,2.5))
plt.title("count of male and female")
sns.countplot(x=gender, hue=gender,legend=["Female", "Male"], fill=False)
plt.show()
```
